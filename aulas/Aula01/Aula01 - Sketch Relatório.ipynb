## Carregando bibliotecas

## Carregando datasets para análise

```python
reclamacoes12 = pd.read_csv(data_path + 'reclamacoes-2012.csv')
reclamacoes13 = pd.read_csv(data_path + 'reclamacoes-2013.csv')
reclamacoes14 = pd.read_csv(data_path +  'reclamacoes-2014.csv')
reclamacoes15 = pd.read_csv(data_path + 'reclamacoes-2015.csv')
reclamacoes16 = pd.read_csv(data_path + 'reclamacoes-2016.csv')

dataset = reclamacoes12.append(reclamacoes13).append(reclamacoes14).append(reclamacoes15).append(reclamacoes16)
```

### Header Metadata

<div class="alert alert-block alert-info">

Tendo confirmado que os datasets estão em um formato padrão e compartilham os nomes dos headers em todos os arquivos, podemos listar a definição de cada coluna:

- **AnoCalendario** : Ano de referência
- **DataArquivamento** : Data de arquivamento da reclamação
- **DataAbertura** : Data de abertura da reclamação
- **CodigoRegiao** : 1 - Norte, 2 - Nordeste, 3 - Sudeste, 4 - Sul, 5 - Centro-Oeste
- **Regiao** : Nome da região
- **UF** : Estado
- **strRazaoSocial** : Razão social da empresa (Base de dados do Procon)
- **strNomeFantasia** : Nome fantasia da empresa (Base de dados do Procon)
- **Tipo** : 1 - Jurídica, 0 - Física
- **NumeroCNPJ** : CNPJ completo
- **RadicalCNPJ** : Primeiros 8 dígitos do CNPJ
- **RazaoSocialRFB** : Razão social da empresa (Base de dados do Governo)
- **NomeFantasiaRFB** : Nome fantasia da empresa (Base de dados do Governo)
- **CNAEPrincipal** : Código de Atividade Econômica da empresa *([Mais informações](https://cnae.ibge.gov.br/))*
- **DescCNAEPrincipal** : Descrição da Atividade Econômica da Empresa
- **Atendida** : Indica se a reclamação foi atendida
- **CodigoAssunto** : Código do assunto da reclamação
- **DescricaoAssunto** : Descrição do assunto da reclamação
- **CodigoProblema** : Código do problema relatado
- **DescricaoProblema** : Descrição do problema relatado
- **SexoConsumidor** : M - Masculino, F - Feminino
- **FaixaEtariaConsumidor** : Faixa etária do consumidor
- **CEPConsumidor** : Código postal do consumidor

## Exploração dos datasets por anos individuais

*Relembrando:* estamos interessados em saber quais empresas são as mais odiadas. Para descobrir isso, começaremos olhando para cada ano individualmente, em ordem cronológica.

### Ano de 2012

#### Avaliação de empresas por região

```python
regions = reclamacoes12['Regiao'].value_counts()
regions
```

```python
norte12    = reclamacoes12.query('CodigoRegiao == 1')
nordeste12 = reclamacoes12.query('CodigoRegiao == 2')
sudeste12  = reclamacoes12.query('CodigoRegiao == 3')
sul12      = reclamacoes12.query('CodigoRegiao == 4')
centro12   = reclamacoes12.query('CodigoRegiao == 5')

regioes  = {'Norte': norte12, 
            'Nordeste': nordeste12, 
            'Sudeste': sudeste12, 
            'Sul': sul12, 
            'Centro-Oeste': centro12
           }
```

```python
empresas = pd.unique(reclamacoes12['RazaoSocialRFB'])

codigos = []
for empresa in empresas:
      codigos.append(reclamacoes12['CodigoRegiao']
             .where(reclamacoes12['RazaoSocialRFB'] == empresa)
             .unique())

plt.scatter(range(len(reclamacoes12['RazaoSocialRFB'].unique())), 
            reclamacoes12['RazaoSocialRFB'].value_counts(dropna=False).get_values(), 
            c=codigos)
```

```python
total_regioes = [len(regiao) for regiao in regioes.values()]

plt.bar(regioes.keys(), total_regioes)
plt.title('Número total de reclamações por região em 2012')
plt.ylabel('Reclamações');
plt.savefig(plot_path + 'total-regioes-2012.png')
```

#### Empresas da região Norte:

#### Avaliação de empresas por setor

```python
rank_setor12 = reclamacoes12['DescCNAEPrincipal'].value_counts()
num_setores_avaliados = 5
```

```python
plt.barh(rank_setor12.keys()[:num_setores_avaliados], rank_setor12[:num_setores_avaliados]);
plt.title('Reclamações por setores em 2012')
plt.xlabel('Reclamações')
plt.savefig(plot_path + 'total-setores-2012.png')
```

```python
rank_empresas_setor12 = []

for setor in rank_setor12.keys()[:num_setores_avaliados]:
    empresas_setor = reclamacoes12['RazaoSocialRFB'].where(reclamacoes12['DescCNAEPrincipal'] == setor).value_counts()
    rank_empresas_setor12.append((setor, empresas_setor))
```

```python
figure, axs = plt.subplots(nrows=num_setores_avaliados, ncols=1, squeeze=False)

for i in range(num_setores_avaliados):
    y = rank_empresas_setor12[i][1].keys()[:num_setores_avaliados]
    x = rank_empresas_setor12[i][1][:num_setores_avaliados]
    setor = rank_empresas_setor12[i][0] 
    axs[i][0].barh(y, x)
    axs[i][0].set_xlabel('Reclamações')
    axs[i][0].set_title('Empresas com maior número de reclamações no setor {} em 2012'.format(setor))

figure.set_figheight(figure.get_figheight()*3)
figure.subplots_adjust(hspace=1)
figure.suptitle('Empresas com maior número de reclamações por setor em 2012')
figure.savefig(plot_path + 'empresas-setores-2012.png');
```
