# Caracterização inicial dos datasets

## Carregando bibliotecas

Vamos carregar as bibliotecas necessárias para trabalhar com as bases de dados.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_path = '../data/'
plot_path = '../plots/'
```

## Carregando bases de dados

Agora podemos ler nossos *datasets* para começarmos nossas análises.

```python
complaints12 = pd.read_csv(data_path + 'reclamacoes-2012.csv')
complaints13 = pd.read_csv(data_path + 'reclamacoes-2013.csv')
complaints14 = pd.read_csv(data_path + 'reclamacoes-2014.csv')
complaints15 = pd.read_csv(data_path + 'reclamacoes-2015.csv')
complaints16 = pd.read_csv(data_path + 'reclamacoes-2016.csv')

dataset = [complaints12, complaints13, complaints14, complaints15, complaints16]
```

## Identificação dos headers de cada dataset

Vamos olhar o nome e tipo das colunas de cada *dataset*, para ter certeza de que estamos com o arquivo correto e de que o mesmo está seguindo a especificação que está no *Kaggle*.

```python
year = 2012
for ds in dataset:
    print('Reclamações de ' + str(year))
    year = year + 1
    ds.info()
    print('----------')
```

## Header Metadata

<div class="alert alert-block alert-info">

Tendo confirmado que os datasets compartilham os nomes dos headers em todos os arquivos, podemos listar a definição de cada coluna:

- **AnoCalendario** : Ano de referência
- **DataArquivamento** : Data de arquivamento da reclamação
- **DataAbertura** : Data de abertura da reclamação
- **CodigoRegiao** : 1 - Norte, 2 - Nordeste, 3 - Sudeste, 4 - Sul, 5 - Centro-Oeste
- **Regiao** : Nome da região
- **UF** : Estado
- **strRazaoSocial** : Razão social da empresa (Base de dados do Procon)
- **strNomeFantasia** : Nome fantasia da empresa (Base de dados do Procon)
- **Tipo** : 1 - Jurídica, 0 - Física
- **NumeroCNPJ** : CNPJ completo
- **RadicalCNPJ** : Primeiros 8 dígitos do CNPJ
- **RazaoSocialRFB** : Razão social da empresa (Base de dados do Governo)
- **NomeFantasiaRFB** : Nome fantasia da empresa (Base de dados do Governo)
- **CNAEPrincipal** : Código de Atividade Econômica da empresa *([Mais informações](https://cnae.ibge.gov.br/))*
- **DescCNAEPrincipal** : Descrição da Atividade Econômica da Empresa
- **Atendida** : Indica se a reclamação foi atendida
- **CodigoAssunto** : Código do assunto da reclamação
- **DescricaoAssunto** : Descrição do assunto da reclamação
- **CodigoProblema** : Código do problema relatado
- **DescricaoProblema** : Descrição do problema relatado
- **SexoConsumidor** : M - Masculino, F - Feminino
- **FaixaEtariaConsumidor** : Faixa etária do consumidor
- **CEPConsumidor** : Código postal do consumidor

# Exploração dos datasets por anos individuais

*Relembrando:* estamos interessados em saber quais empresas são as mais odiadas. Para descobrir isso, começaremos olhando para cada ano individualmente, em ordem cronológica.

## Ano de 2012

### Avaliação por região

```python
regions = complaints12['Regiao'].value_counts()
regions.describe()
```

```python
plt.bar(regions.keys(), regions)
plt.title('Número total de reclamações por região em 2012')
plt.ylabel('Reclamações');
plt.savefig(plot_path + 'regioes-2012.png')
```

#### Região Norte

```python
rank_norte = complaints12['RazaoSocialRFB'].where(complaints12['CodigoRegiao']==1).value_counts()
rank_norte.describe()
```

```python
plt.barh(rank_norte.keys()[:10], rank_norte.get_values()[:10])
plt.title('Empresas com maior número de reclamações na região Norte em 2012')
plt.xlabel('Reclamações');
plt.savefig(plot_path + 'norte-2012.png')
```

#### Região Nordeste

```python
rank_nordeste = complaints12['RazaoSocialRFB'].where(complaints12['CodigoRegiao']==2).value_counts()
rank_nordeste.describe()
```

```python
plt.barh(rank_nordeste.keys()[:10], rank_nordeste.get_values()[:10])
plt.title('Empresas com maior número de reclamações na região Nordeste em 2012')
plt.xlabel('Reclamações');
plt.savefig(plot_path + 'nordeste-2012.png')
```

#### Região Sudeste

```python
rank_sudeste = complaints12['RazaoSocialRFB'].where(complaints12['CodigoRegiao']==3).value_counts()
rank_sudeste.describe()
```

```python
plt.barh(rank_sudeste.keys()[:10], rank_sudeste.get_values()[:10])
plt.title('Empresas com maior número de reclamações na região Sudeste em 2012')
plt.xlabel('Reclamações');
plt.savefig(plot_path + 'sudeste-2012.png')
```

#### Região Sul

```python
rank_sul = complaints12['RazaoSocialRFB'].where(complaints12['CodigoRegiao']==4).value_counts()
rank_sul.describe()
```

#### Região Centro-Oeste

### Avaliação por setor

```python
rank_setor = complaints12['DescCNAEPrincipal'].value_counts()
rank_setor.describe()
```

```python
plt.barh(rank_setor.keys()[:5], rank_setor.get_values()[:5])
plt.title('Número total de reclamações por setor em 2012')
plt.xlabel('Reclamações')
plt.savefig(plot_path + 'setores-2012.png')
```

```python
rank_empresas_setor = []

for setor in rank_setor.keys():
    empresas_setor = complaints12['RazaoSocialRFB'].where(complaints12['DescCNAEPrincipal'] == setor).value_counts()
    rank_empresas_setor.append((setor, empresas_setor))
```

```python
figure, axs = plt.subplots(nrows=5, ncols=1, squeeze=False)

for i in range(5):
    y = rank_empresas_setor[i][1].keys()[:5]
    x = rank_empresas_setor[i][1].get_values()[:5]
    setor = rank_empresas_setor[i][0]
    
    axs[i][0].barh(y, x)
    axs[i][0].set_xlabel('Reclamações')
    axs[i][0].set_title('Empresas com maior número de reclamações no setor {} em 2012'.format(setor))

figure.set_figheight(figure.get_figheight()*3)
figure.subplots_adjust(hspace=1)
figure.suptitle('Empresas com maior número de reclamações por setor em 2012')
figure.savefig(plot_path + 'empresas-setores-2012.png');
```
